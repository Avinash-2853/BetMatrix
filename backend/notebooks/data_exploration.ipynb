```python
import requests
from functools import lru_cache

@lru_cache(maxsize=32)
def get_coach_name(team_id, year):
    """
    Fetch head coach's first and last name for a given team ID and season.
    """
    url = f"https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/seasons/{year}/teams/{team_id}/coaches"
    resp = requests.get(url).json()
    
    # Get the first coach's detail (usually head coach)
    coach_details = requests.get(resp["items"][0]["$ref"]).json()
    return coach_details.get("firstName"), coach_details.get("lastName")


def get_team_details(year: int, week: int):
    """
    Fetch NFL schedule details including:
    - home team abbreviation
    - away team abbreviation
    - head coach first and last name
    - stadium
    """
    url = f"https://cdn.espn.com/core/nfl/schedule?xhr=1&year={year}&week={week}"
    response = requests.get(url)
    data = response.json()

    games = []
    schedule = data.get("content", {}).get("schedule", {})

    for date_key, date_data in schedule.items():
        for game in date_data.get("games", []):
            comp = game["competitions"][0]
            teams = comp["competitors"]

            home = next((t for t in teams if t.get("homeAway") == "home"), None)
            away = next((t for t in teams if t.get("homeAway") == "away"), None)
            if not home or not away:
                continue

            home_abbr = home["team"].get("abbreviation")
            away_abbr = away["team"].get("abbreviation")
            stadium = comp.get("venue", {}).get("fullName")

            home_id = home["team"].get("id")
            away_id = away["team"].get("id")

            # Fetch head coach names
            home_first, home_last = get_coach_name(home_id, year)
            away_first, away_last = get_coach_name(away_id, year)

            games.append({
                "home_team": home_abbr,
                "home_coach": f"{home_first} {home_last}",
                "away_team": away_abbr,
                "away_coach": f"{away_first} {away_last}",
                "stadium": stadium
            })

    return games




```

```python
import requests

def get_match_winner(event_id):
    """Fetch the winning team's abbreviation for a given NFL event ID from ESPN's API."""
    base_url = f"https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/events/{event_id}"
    
    # Get event data
    event_response = requests.get(base_url)
    event_data = event_response.json()
    
    # Get competition details
    competition_url = event_data.get("competitions", [])[0]["$ref"]
    competition_data = requests.get(competition_url).json()
    
    # Get list of competitors
    competitors = competition_data["competitors"]
    
    winner_team_abbr = None
    
    # Loop through competitors to find the winner
    for team_info in competitors:
        team_data = requests.get(team_info["$ref"]).json()
        if team_data.get("winner", False):
            # Fetch team details (to get abbreviation)
            team_url = team_data["team"]["$ref"]
            team_details = requests.get(team_url).json()
            winner_team_abbr = team_details["abbreviation"]
            break
    
    return winner_team_abbr

# Example
print(get_match_winner("401772937"))
```

```python
import requests
from functools import lru_cache
from typing import Optional, Tuple, List, Dict, Any


@lru_cache(maxsize=32)
def get_coach_name(team_id: str, year: int) -> Tuple[Optional[str], Optional[str]]:
    """
    Fetch head coach's first and last name for a given team ID and season.
    Returns:
        (first_name, last_name) or (None, None) if not found.
    """
    url = f"https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/seasons/{year}/teams/{team_id}/coaches"
    resp = requests.get(url).json()

    # Defensive coding: handle missing or malformed data
    try:
        coach_details = requests.get(resp["items"][0]["$ref"]).json()
        return coach_details.get("firstName"), coach_details.get("lastName")
    except (KeyError, IndexError, requests.RequestException):
        return None, None


def get_team_logo(team_id: str, year: int) -> Optional[str]:
    """
    Fetch team logo URL from ESPN team API.
    Args:
        team_id (str): Team ID
        year (int): Season year
    Returns:
        str | None: Team logo URL if available, else None
    """
    try:
        url = f"http://sports.core.api.espn.com/v2/sports/football/leagues/nfl/seasons/{year}/teams/{team_id}?lang=en&region=us"
        response = requests.get(url)
        data = response.json()

        logos = data.get("logos", [])
        if logos and isinstance(logos, list):
            return logos[0].get("href")
    except requests.RequestException:
        pass

    return None


def map_team(team_name: str) -> str:
    """ 
    Map team abbreviations to standard forms.
    """
    if team_name == "WSH":
        return "WAS"
    if team_name == "LAR":
        return "LA"
    return team_name


def get_team_details(year: int, week: int) -> List[Dict[str, Any]]:
    """
    Fetch NFL schedule details including:
    - home team abbreviation
    - away team abbreviation
    - head coach names
    - stadium
    - game ID
    - home and away team logo URLs

    Returns:
        List[Dict]: A list of game dictionaries with all details.
    """
    url = f"https://cdn.espn.com/core/nfl/schedule?xhr=1&year={year}&week={week}"
    response = requests.get(url)
    data = response.json()

    games: List[Dict[str, Any]] = []
    schedule = data.get("content", {}).get("schedule", {})

    for date_key, date_data in schedule.items():
        for game in date_data.get("games", []):
            comp = game["competitions"][0]
            teams = comp["competitors"]

            home = next((t for t in teams if t.get("homeAway") == "home"), None)
            away = next((t for t in teams if t.get("homeAway") == "away"), None)
            if not home or not away:
                continue

            home_abbr = home["team"].get("abbreviation")
            away_abbr = away["team"].get("abbreviation")
            stadium = comp.get("venue", {}).get("fullName")
            game_id = comp.get("id")

            home_id = home["team"].get("id")
            away_id = away["team"].get("id")

            home_logo = get_team_logo(home_id, year)
            away_logo = get_team_logo(away_id, year)

            home_first, home_last = get_coach_name(home_id, year)
            away_first, away_last = get_coach_name(away_id, year)

            games.append({
                "game_id": game_id,
                "home_team": map_team(home_abbr),
                "home_coach": f"{home_first or ''} {home_last or ''}".strip(),
                "home_team_logo_url": home_logo,
                "away_team": map_team(away_abbr),
                "away_coach": f"{away_first or ''} {away_last or ''}".strip(),
                "away_team_logo_url": away_logo,
                "stadium": stadium
            })

    return games


def get_match_winner(event_id: str) -> Optional[str]:
    """
    Fetch the winning team's abbreviation for a given NFL event ID from ESPN's API.
    
    Returns:
        str | None: The winner's abbreviation if found, else None
    """
    base_url = f"https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/events/{event_id}"

    try:
        event_response = requests.get(base_url)
        event_data = event_response.json()

        competitions = event_data.get("competitions", [])
        if not competitions or not isinstance(competitions, list):
            return None

        competition_url = competitions[0].get("$ref")
        if not competition_url:
            return None

        competition_data = requests.get(competition_url).json()
        competitors = competition_data.get("competitors", [])
        if not competitors:
            return None

        for team_info in competitors:
            team_data = requests.get(team_info["$ref"]).json()
            if team_data.get("winner", False):
                team_url = team_data["team"]["$ref"]
                team_details = requests.get(team_url).json()
                return team_details.get("abbreviation")

    except requests.RequestException:
        pass

    return None
```

```python
import requests
from typing import Optional, Dict

def get_match_scores(event_id: str) -> Optional[Dict[str, int]]:
    """
    Fetches home and away team scores for a given NFL event ID from ESPN's API.
    
    This corrected version avoids redundant API calls and correctly follows the '$ref'
    links to retrieve the final scores for each team.

    Args:
        event_id: The unique identifier for the ESPN event (e.g., "401220116").

    Returns:
        A dictionary with "home_score" and "away_score", or None if an error occurs.
        Example: {"home_score": 27, "away_score": 17}
    """
    # The primary URL for the event data
    base_url = f"https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/events/{event_id}"
    
    try:
        # Step 1: Fetch the main event data in a single request.
        event_data = requests.get(base_url).json()

        # Step 2: Navigate directly to the list of competitors.
        # The structure is event -> competitions[0] -> competitors
        competitors = event_data.get("competitions", [{}])[0].get("competitors", [])

        if not competitors or len(competitors) != 2:
            print("Error: Competitors data not found or incomplete.")
            return None

        scores = {}

        # Step 3: Loop through the two competitors to get their scores.
        for competitor in competitors:
            team_side = competitor.get("homeAway")
            
            # The 'score' key contains a dictionary with a '$ref' URL to the actual score data.
            # We must fetch this URL to get the score value.
            score_ref_url = competitor.get("score", {}).get("$ref")
            
            if not team_side or not score_ref_url:
                continue # Skip if essential data is missing

            # Step 4: Fetch the score from its specific endpoint and get the 'value'.
            score_data = requests.get(score_ref_url).json()
            score_value = int(score_data.get("value", 0))

            if team_side == "home":
                scores["home_score"] = score_value
            elif team_side == "away":
                scores["away_score"] = score_value
        
        # Ensure both scores were found before returning
        if "home_score" in scores and "away_score" in scores:
            return scores
        else:
            print("Error: Could not determine both home and away scores.")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from ESPN API: {e}")
        return None
    except (KeyError, IndexError, TypeError) as e:
        print(f"Error parsing the JSON data structure: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
```
